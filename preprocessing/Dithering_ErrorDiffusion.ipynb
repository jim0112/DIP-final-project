#Dithering

```python
def double(img):
    out = np.zeros((img.shape[0]*2, img.shape[1]*2))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            out[i,j] = 4 * img[i,j] + 1
            out[i, j + img.shape[1]] = 4 * img[i,j] + 2 
            out[i + img.shape[0], j] = 4 * img[i,j] + 3 
            out[i + img.shape[0], j + img.shape[1]] =  4 * img[i,j]
    return out

def gen_thres(size):
    out = [[1,2],[3,0]]
    out = np.array(out)
    while size > 1:
        out = double(out)
        size -= 1
    
    thres_mat = np.zeros((out.shape[0], out.shape[1]))
    N = out.shape[0]
    for i in range(out.shape[0]):
        for j in range(out.shape[1]):
            thres_mat[i,j] = 255 * (out[i,j] + 0.5) / N**2
    return thres_mat

def dithering(img, size):
    thres_mat = gen_thres(size)
    fin_out = np.zeros((img.shape[0], img.shape[1], img.shape[2]))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            for k in range(img.shape[2]):
                if img[i,j,k] > thres_mat[i][j]:
                    fin_out[i,j,k] = 255
                else:
                    fin_out[i,j,k] = 0
    return fin_out
```

```python
import cv2
import os
import numpy as np
def findAllFile_dithering(base, aim):
    count = 0
    for root, ds, fs in os.walk(base):
        for f in fs:
            if f.endswith('.png'):
              fullname = os.path.join(root, f)
              img = cv2.imread(fullname)
              img = np.array(img)
              out = dithering(img,5)
              outname = os.path.join(aim, f)
              cv2.imwrite(outname,out)
              count+=1
              if count % 100 == 0:
                print(count,outname)
```

#Error Diffusion

```python
def error_diffusion_dithering(img,mask):
    
    h, w = img.shape

    F = np.zeros((h, w))
    for i in range(h):
        for j in range(w):
            F[i,j] = img[i,j]

    threshold = 127

    for y in range(h):
        for x in range(w):
            old = F[y][x]
            if F[y][x] < threshold:
                F[y,x] = 0
            else:
                F[y,x] = 255
            error =  old - F[y,x]

            for dx, dy, c in mask:
                xn, yn = x + dx, y + dy
                if (0 <= xn < h) and (0 <= yn < w):
                    F[yn][xn] += error * c
    
    return F.astype(np.uint8)



floyd_mask =  (
    (1, 0, 7 / 16),
    (-1, 1, 3 / 16),
    (0, 1, 5 / 16),
    (1, 1, 1 / 16),
)

jarvis_mask = (
    (1, 0, 7 / 48),
    (2, 0, 5 / 42),
    (-2, 1, 3 / 48),
    (-1, 1, 5 / 48),
    (0, 1, 7 / 48),
    (1, 1, 5 / 48),
    (2, 1, 3 / 48),
    (-2, 2, 1 / 48),
    (-1, 2, 3 / 48),
    (0, 2, 5 / 48),
    (1, 2, 3 / 48),
    (2, 2, 1 / 48),
)

stucki_map =  (
        (1, 0, 8 / 42),
        (2, 0, 4 / 42),
        (-2, 1, 2 / 42),
        (-1, 1, 4 / 42),
        (0, 1, 8 / 42),
        (1, 1, 4 / 42),
        (2, 1, 2 / 42),
        (-2, 2, 1 / 42),
        (-1, 2, 2 / 42),
        (0, 2, 4 / 42),
        (1, 2, 2 / 42),
        (2, 2, 1 / 42),
)
def error_diffusion_3_channel(img):
    for i in range(1):
        out0 = error_diffusion_dithering(img[:,:,0],floyd_mask)
        out1 = error_diffusion_dithering(img[:,:,1],floyd_mask)
        out2 = error_diffusion_dithering(img[:,:,2],floyd_mask)
    out0 = np.reshape(out0,(32,32,1))
    out1 = np.reshape(out1,(32,32,1))
    out2 = np.reshape(out2,(32,32,1))
    out = np.concatenate((out0,out1,out2),axis = 2)
    return out
```

```python
import cv2
import os
import numpy as np
def findAllFile(base, aim):
    count = 0
    for root, ds, fs in os.walk(base):
        for f in fs:
            if f.endswith('.png'):
              fullname = os.path.join(root, f)
              img = cv2.imread(fullname)
              img = np.array(img)
              out = error_diffusion_3_channel(img)
              outname = os.path.join(aim, f)
              cv2.imwrite(outname,out)
              count+=1
              if count % 100 == 0:
                print(count,outname)
```
